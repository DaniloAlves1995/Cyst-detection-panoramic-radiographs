# Automatic cyst detection on radiograph images


This process is an experiment of a larger methodology that aims to automatically detect dental cysts in panoramic radiography images. In this experiment, the U-Net convolutional neural network (CNN) is applied to detect cysts through pixel segmentation.

The process of segmenting an image consists of separating it into regions so that an objective region and a background region can be identified. This procedure is done by U-Net, which is used to classify each pixel in two regions: cyst region or background.

**dataset:** A Data augmentation technique was applied to the radiography dataset in another algorithm, generating 40 images for each exam, totaling 1279 images.

**validation:** The cross-validation used in this experiment is the Leave-one-out, which consists of selecting one exam from the dataset to be used as a test group and the others as a training group, and repeating this iteratively until all exams have participated as a group. of test.

## Functions for read and processing
These functions are used to read, apply pre-processing to the images and prepare the data pipeline to feed the model with the training, testing and validation groups. The database used is already pre-processing by another algorithm, for now we just apply resizing as pre-processing. 

Resizing is not mandatory because U-Net is a fully convolutional network and does not depend on the size of the input. Even so, we chose it due to memory constraints. Also, if the images have not been resized it is necessary to use batch size 1 as it is not possible to group images with different sizes in the same batch.

```python
def load_data_from_zip(file):
    """ Function to unzip files """
    with zipfile.ZipFile(file, "r") as zip_ref:
        unzipped_file = zip_ref.namelist()[0]
        zip_ref.extractall()

def _process_pathnames(fname, label_path):
    """ Read and process each image to form the dataset. 
        We map this function onto each pathname pair.
    """
  
    img_str = tf.io.read_file(fname)
    img = tf.image.decode_jpeg(img_str, channels=3)

    label_img_str = tf.io.read_file(label_path)
    # They return as (num_frames, h, w, c)
    label_img = tf.image.decode_png(label_img_str)[0]
    # The label have values of 1 or 0, indicating pixel wise
    # cyst or not (background). 
    label_img = tf.expand_dims(label_img, axis=-1)
    return img, label_img

def _augment(img, label_img, resize=None, scale=1): 
    """ Apply resize techinique for exam image and label image """
    if resize is not None:
        label_img = tf.image.resize(label_img, resize)
        img = tf.image.resize(img, resize)
  
    label_img = tf.cast(label_img, dtype=tf.float32) * scale
    img = tf.cast(img, dtype=tf.float32) * scale 
    return img, label_img

def get_baseline_dataset(filenames, 
                         labels,
                         preproc_fn=functools.partial(_augment),
                         threads=5, 
                         batch_size=batch_size,
                         shuffle=True):        
    """Gets a data pipeline for apply on the model from files names of groups"""   
    num_x = len(filenames)
    # Create a dataset using both images and labels
    dataset = tf.data.Dataset.from_tensor_slices((filenames, labels))
    # Apply preprocessing function to all in the dataset, using
    # multithreading
    dataset = dataset.map(_process_pathnames, num_parallel_calls=threads)
    if preproc_fn.keywords is not None and 'resize' not in preproc_fn.keywords:
    assert batch_size == 1, "Batching images don't have the same size"

    dataset = dataset.map(preproc_fn, num_parallel_calls=threads)

    if shuffle:
    dataset = dataset.shuffle(num_x)

    # Repeat for all epochs 
    dataset = dataset.repeat().batch(batch_size)
    return dataset
```

```python
# Unzip folder
load_data_from_zip('exames.zip')
load_data_from_zip('masks.zip')

# Load the file paths
img_dir = "exames"
label_dir = "masks"

x_train = [os.path.join(img_dir, f) for f in os.listdir(img_dir) if os.path.isfile(os.path.join(img_dir, f))]
y_train = [os.path.join(label_dir, f) for f in os.listdir(label_dir) if os.path.isfile(os.path.join(label_dir, f))]

# Sort the arrays
x_train.sort()
y_train.sort()

# Preprocessing used
tr_cfg = {
      'resize': [img_shape[0], img_shape[1]],
      'scale': 1 / 255.
}
tr_preprocessing_fn = functools.partial(_augment, **tr_cfg)
```

## U-Net model
The model is built using two parts, one to enconder and another to decoder. The encoder portion is composed of a linear stack of Conv, BatchNorm, and Relu operations followed by a MaxPool. The Decoder portion is comprised of UpSampling2D, Conv, BatchNorm, and Relus. We used a concatenate the feature map of the same size on the decoder side and add batch normalization to each of blocks.

```python
def conv_block(input_tensor, num_filters):
    encoder = layers.Conv2D(num_filters, (3, 3), padding='same')(input_tensor)
    encoder = layers.BatchNormalization()(encoder)
    encoder = layers.Activation('relu')(encoder)
    encoder = layers.Conv2D(num_filters, (3, 3), padding='same')(encoder)
    encoder = layers.BatchNormalization()(encoder)
    encoder = layers.Activation('relu')(encoder)
    return encoder

def encoder_block(input_tensor, num_filters):
    encoder = conv_block(input_tensor, num_filters)
    encoder_pool = layers.MaxPooling2D((2, 2), strides=(2, 2))(encoder)

    return encoder_pool, encoder

def decoder_block(input_tensor, concat_tensor, num_filters):
    decoder = layers.Conv2DTranspose(num_filters, (2, 2), strides=(2, 2), padding='same')(input_tensor)
    decoder = layers.concatenate([concat_tensor, decoder], axis=-1)
    decoder = layers.BatchNormalization()(decoder)
    decoder = layers.Activation('relu')(decoder)
    decoder = layers.Conv2D(num_filters, (3, 3), padding='same')(decoder)
    decoder = layers.BatchNormalization()(decoder)
    decoder = layers.Activation('relu')(decoder)
    decoder = layers.Conv2D(num_filters, (3, 3), padding='same')(decoder)
    decoder = layers.BatchNormalization()(decoder)
    decoder = layers.Activation('relu')(decoder)
    return decoder
```

```python
# Building the architecture

inputs = layers.Input(shape=img_shape)
# 256

encoder0_pool, encoder0 = encoder_block(inputs, 32)
# 128

encoder1_pool, encoder1 = encoder_block(encoder0_pool, 64)
# 64

encoder2_pool, encoder2 = encoder_block(encoder1_pool, 128)
# 32

encoder3_pool, encoder3 = encoder_block(encoder2_pool, 256)
# 16

encoder4_pool, encoder4 = encoder_block(encoder3_pool, 512)
# 8

center = conv_block(encoder4_pool, 1024)
# center

decoder4 = decoder_block(center, encoder4, 512)
# 16

decoder3 = decoder_block(decoder4, encoder3, 256)
# 32

decoder2 = decoder_block(decoder3, encoder2, 128)
# 64

decoder1 = decoder_block(decoder2, encoder1, 64)
# 128

decoder0 = decoder_block(decoder1, encoder0, 32)
# 256

outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(decoder0)
```

```python
# Define the model by specifying the inputs and outputs associated  
model = models.Model(inputs=[inputs], outputs=[outputs])
```

### Metrics and loss functions
This metric measure the overlap between two regions and what we're looking for is maximizing that. Also, we chose this metric because it performs better on class-unbalanced problems by design. Using cross entropy is more of a proxy which is easier to maximize. Instead, we maximize our objective directly. 

```python
def dice_coeff(y_true, y_pred):
    smooth = 1.

    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    score = (2. * intersection + smooth) / (tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth)
    return score
  
def dice_loss(y_true, y_pred):
    loss = 1 - dice_coeff(y_true, y_pred)
    return loss

def bce_dice_loss(y_true, y_pred):
    loss = losses.binary_crossentropy(y_true, y_pred) + dice_loss(y_true, y_pred)
    return loss  
```

## Executing the process
Starts the cross-validation process with the model. For each iteration, one image from a group exam is taken and used as a group test. This image and all other generated from the same exam through the argumentation process are removed. After training, we save the history and load the best saved model to apply the assessment with the test set. We save the evaluation metrics in a file and three images (input image, input mask and output prediction). All these steps are done for each exam.

```python

for num_img in range(0,32):

    print("Process with exam "+str(num_img)+":\n")

    # Separate an image for test set
    x_img_test = [x_train[num_img*40], x_train[num_img*40]]
    y_img_test = [y_train[num_img*40], y_train[num_img*40]]

    # Remove all images wich were genereted for the same exam used for test set
    x_train_filenames = x_train[0:num_img*40] + x_train[(num_img+1)*40:len(x_train)]
    y_train_filenames = y_train[0:num_img*40] + y_train[(num_img+1)*40:len(y_train)]

    # Separate part of train set to used as validation during the train
    x_train_filenames, x_val_filenames, y_train_filenames, y_val_filenames = \
                      train_test_split(x_train_filenames, y_train_filenames, test_size=0.2, random_state=42)

    # train set size and validation
    num_train_examples = len(x_train_filenames)
    num_val_examples = len(x_val_filenames)

    tr_preprocessing_fn = functools.partial(_augment, **tr_cfg)

    # Create pipeline for train set
    train_ds = get_baseline_dataset(x_train_filenames,
                                  y_train_filenames,
                                  preproc_fn=tr_preprocessing_fn,
                                  batch_size=batch_size)

    # Create pipeline for validation set
    val_ds = get_baseline_dataset(x_val_filenames,
                                y_val_filenames, 
                                preproc_fn=tr_preprocessing_fn,
                                batch_size=batch_size)

    # Create pipeline fot test set
    teste_ds = get_baseline_dataset(x_img_test,
                                  y_img_test, 
                                  preproc_fn=tr_preprocessing_fn,
                                  batch_size=batch_size)

    # Compile the model
    model.compile(optimizer='adam', loss=bce_dice_loss, metrics=[dice_loss, dice_coeff])

    # Configure to salve the best weights
    save_model_path = 'weights_'+str(num_img)+'.hdf5'
    cp = tf.keras.callbacks.ModelCheckpoint(filepath=save_model_path, monitor='val_dice_loss', save_best_only=True)

    # Configure stop rule
    es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=7)

    # Start the training
    history = model.fit(train_ds, 
                     steps_per_epoch=int(np.ceil(num_train_examples / float(batch_size))),
                     epochs=epochs,
                     validation_data=val_ds,
                     validation_steps=int(np.ceil(num_val_examples / float(batch_size))),
                     callbacks=[cp,es])

    # Save train history
    with open('/history_train_'+str(num_img), 'wb') as file_pi:
        pickle.dump(history.history, file_pi)

    # Load best model saved
    model = []
    model = models.load_model(save_model_path, custom_objects={'bce_dice_loss': bce_dice_loss,
                                                           'dice_loss': dice_loss, 'dice_coeff': dice_coeff})

    # Evaluate with test pipeline
    test_metric = model.evaluate(teste_ds, steps=200, verbose=0)

    # Salve the metrics got from test image [loss, dice_loss, dice_coeff]
    with open("metrics.txt", "a") as f:
        f.write(""+str(test_metric[0])+" "+str(test_metric[1])+" "+str(test_metric[2])+"\n")

    # Get predict image of this iteration
    data_aug_iter = teste_ds.make_one_shot_iterator()
    next_element = data_aug_iter.get_next()

    batch_of_imgs, label = tf.keras.backend.get_session().run(next_element)
    img = batch_of_imgs[0]
    predicted_label = model.predict(batch_of_imgs, verbose=0)[0]

    # Clean variables
    model = []
    train_ds = []
    val_ds = []
    teste_ds = []

    # Save images of this iteration
    !mkdir saidas
    save_img("outs/img_"+str(num_img)+".tiff", img)

    mask = PIL.Image.fromarray(label[0, :, :, 0])
    mask.save("outs/mask_"+str(num_img)+".tiff", format="TIFF")

    pred = PIL.Image.fromarray(predicted_label[:, :, 0])
    pred.save("out/pred_"+str(num_img)+".tiff", format="TIFF")
  
```
